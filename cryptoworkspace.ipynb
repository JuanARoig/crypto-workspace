```
import builtins
def gcd(a, b):
  while b != 0:
    t = b
    b = a%b
    a = t
  return a
```

```
gcd(6, 9)
```

```
def egcd(a, b):
  x = 0
  y = 1
  u = 1
  v = 0
  while a != 0:
    q = b//a
    r = b%a
    m = x - u*q
    n = y - v*q
    b = a
    a = r
    x = u
    y = v
    u = m
    v = n
  gcd = b
  return gcd, x, y
```

```
egcd(3, 5)
```

```
from random import randint

a = 288260533169915
p = 1007621497415251

FLAG = b'crypto{????????????????????}'


def encrypt_flag(flag):
    ciphertext = []
    plaintext = ''.join([bin(i)[2:].zfill(8) for i in flag])
    for b in plaintext:
        print(b)
        e = randint(1, p)
        n = pow(a, e, p)
        if b == '1':
            ciphertext.append(n)
        else:
            n = -n % p
            ciphertext.append(n)
    return ciphertext

encFLAG = encrypt_flag(FLAG)
print(encFLAG)
```
