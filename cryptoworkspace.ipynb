```
import builtins
def gcd(a, b):
  while b != 0:
    t = b
    b = a%b
    a = t
  return a
```

```
def egcd(a, b):
  x = 0
  y = 1
  u = 1
  v = 0
  while a != 0:
    q = b//a
    r = b%a
    m = x - u*q
    n = y - v*q
    b = a
    a = r
    x = u
    y = v
    u = m
    v = n
  gcd = b
  return gcd, x, y
```
